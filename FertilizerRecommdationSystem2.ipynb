```python
from sklearn.preprocessing import LabelEncoder

# Load dataset
fert_df = pd.read_csv('/Fertilizer Prediction.csv')

# Identify and encode categorical columns
categorical_cols = ['Soil Type', 'Crop Type']  # Replace with actual categorical column names in your CSV

le_dict = {}  # To keep track of encoders if needed later
for col in categorical_cols:
    le = LabelEncoder()
    fert_df[col] = le.fit_transform(fert_df[col])
    le_dict[col] = le  # Save encoder if you need to inverse-transform later

# Features and Target
X_fert = fert_df.drop('Fertilizer Name', axis=1)
y_fert = fert_df['Fertilizer Name']

# Encode target
le_fert = LabelEncoder()
y_fert_encoded = le_fert.fit_transform(y_fert)

# Split the dataset
from sklearn.model_selection import train_test_split
Xf_train, Xf_test, yf_train, yf_test = train_test_split(X_fert, y_fert_encoded, test_size=0.2, random_state=42)

# Scale
from sklearn.preprocessing import StandardScaler
scaler_fert = StandardScaler()
Xf_train_scaled = scaler_fert.fit_transform(Xf_train)
Xf_test_scaled = scaler_fert.transform(Xf_test)

# --- Support Vector Machine ---
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm_fert = SVC()
svm_fert.fit(Xf_train_scaled, yf_train)
fert_svm_pred = svm_fert.predict(Xf_test_scaled)
print("SVM Accuracy (Fertilizer):", round(accuracy_score(yf_test, fert_svm_pred) * 100, 2), "%")

# --- Decision Tree ---
from sklearn.tree import DecisionTreeClassifier

tree_fert = DecisionTreeClassifier()
tree_fert.fit(Xf_train, yf_train)  # No need to scale for trees
fert_tree_pred = tree_fert.predict(Xf_test)
print("Decision Tree Accuracy (Fertilizer):", round(accuracy_score(yf_test, fert_tree_pred) * 100, 2), "%")
```

```python

from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
# Load data
fertilizer = pd.read_csv("/Fertilizer Prediction.csv")

# Clean column names
fertilizer.columns = fertilizer.columns.str.strip()  # Remove leading/trailing spaces
fertilizer.columns = fertilizer.columns.str.lower()  # Optional: convert to lowercase

# Check column names
print(fertilizer.columns)

# Split X and y
X = fertilizer.drop('fertilizer name', axis=1)
y = fertilizer['fertilizer name']

# Identify categorical and numeric columns
categorical_cols = ['soil type', 'crop type']  # Adjust based on cleaned column names
numeric_cols = [col for col in X.columns if col not in categorical_cols]

# Preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', MinMaxScaler(), numeric_cols)
    ]
)

# Combine preprocessing and classifier in a pipeline
pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))  # Replace with any classifier
])

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the pipeline
pipeline.fit(X_train, y_train)

# Evaluate
accuracy = pipeline.score(X_test, y_test)
print(f"Accuracy: {round(accuracy * 100, 2)}%")
```
